# Transpose row to columns

```python
import pandas as pd
```

```python
path_to_raw_data="H:/IR&A Technical Students/Saumya/IPEDS_03/1. Raw Data"
df=pd.read_csv(path_to_raw_data+"\\Fall Enrollment\\"+"ef2013b_rv.csv")
```

```python
df
```

```python
df=df[(df["LSTUDY"]==2)]
```

```python
df=df.loc[:,~df.columns.str.startswith('X')]
```

```python
df=df.drop(["LSTUDY","LINE"],axis=1)
```

```python
df=df.set_index("UNITID")
```

```python
df
```

```python
list_df = []
```

```python
for i in range(1,15):
    list_df.append(df[df['EFBAGE'] == i])
```

```python
list_df
```

```python
df[df['EFBAGE'] == 1]
```

```python
descr = ['_All', '_AgeUnder25', '_AgeUnder18', '_Age18-19', '_Age20-21', '_Age21-24', '_Age25&OverTotal', '_Age25-29', 
         '_Age30-34', '_Age35-39', '_Age40-49', '_Age50-64', '_Age65&over', '_AgeUnknown']
```

```python
for i in range(0,14):
    list_df[i].columns = [str(col) + descr[i]+'_2013'  for col in list_df[i].columns]
```

```python
final = pd.concat(list_df, axis=1)
```

```python
final
```

```python
final.to_csv("H:/IR&A Technical Students/Saumya/IPEDS_03/2. Processed data/ef2013.csv")
```
