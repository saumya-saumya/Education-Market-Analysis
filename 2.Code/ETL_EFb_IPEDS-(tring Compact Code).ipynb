```python
fall_enroll_dict = {re.findall(r'\d+',filenames[i].split('\\')[-1])[0]:pd.read_csv(filenames[i]) for i in range(len(filenames))}
```

```python
for fileyr, df in fall_enroll_dict.items():
    fall_enroll_dict[fileyr] = df[df["LSTUDY"] == 2].loc[:,~df.columns.str.startswith('X')].drop(["LSTUDY","LINE"],axis=1).set_index("UNITID")
```

```python
descr = ['_All', '_AgeUnder25', '_AgeUnder18', '_Age18-19', '_Age20-21', '_Age21-24', '_Age25&OverTotal', '_Age25-29', 
         '_Age30-34', '_Age35-39', '_Age40-49', '_Age50-64', '_Age65&over', '_AgeUnknown']
result_dict = {descr[i]: {} for i in range(len(descr))}
```
